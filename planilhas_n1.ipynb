Este dataset oferece uma rica base de dados para análise de filmes, desde suas características básicas até como os usuários interagem com eles, por meio de avaliações e tags.


1B) Cada linha da planilha MOVIE descreve um filme específico, com seu título e os gêneros associados. O movieId é a chave para vincular esta informação a outras planilhas, como os ratings e tags.

Cada linha da planilha RATINGS representa uma avaliação de um filme por um usuário específico. O userId e o movieId formam um par que indica qual usuário avaliou qual filme, e o rating é a nota dada ao filme.

cada linha da planilha TAGS representa uma tag associada a um filme específico, dada por um usuário. O userId e o movieId mostram quem e qual filme foi marcado com a tag, e a tag descreve algo sobre o filme, como seu gênero ou uma característica adicional.

Cada linha da planilha Link representa um único filme, com os identificadores únicos para os sites IMDb e TMDb.


1C) A planilha LINK relaciona os filmes a links específicos de outras plataformas (IMDb, TMDb). O tipo de dado int64 é usado para os IDs numéricos, e o str é possível para casos em que os identificadores podem ser alfanuméricos (como no IMDb).

A planilha FILME contém informações descritivas sobre os filmes. O movieId é a chave para vincular com outras planilhas, e title e genres são do tipo string (str), onde o título é uma string representando o nome do filme, e genres é uma string que contém os gêneros separados por |.

A planilha de RATINGS contém as avaliações feitas pelos usuários. O userId e movieId são inteiros, enquanto rating é um número de ponto flutuante (float64), pois pode haver decimais. O timestamp é um valor inteiro representando o tempo exato da avaliação (em formato Unix).

A planilha de TAGS contém palavras-chave associadas a filmes, atribuídas por usuários. O userId e movieId são inteiros, enquanto o tag é uma string representando a palavra-chave e o timestamp é um inteiro representando o tempo em que a tag foi adicionada.
